### *Being able to produce graphs and plots is not only important for use in scientific papers, etc., but also being able to quickly visualize data properly will save you a lot of time when working with large data sets.*

**Matplotlib:** https://matplotlib.org/stable/index.html

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#kriva sjaja promenljive zvezde, tj. promena magnitude u toku vremena

t = np.linspace(0, 10, 500) #500 ekvidistantnih tačaka od 0 do 10 dana 

mag = 12 + 0.3*np.sin(2*np.pi*t) # promena magnitude je opisana ovom funkcijom

plt.figure() # plotuje praznu figuru
# prvo navodim x pa y osu kada plotujem
plt.plot(t, mag, c='red', linestyle='--', label='Kriva sjaja') # plotuje podatke i povezuje ih linijom, mogu da biram boju, stil linije, dodam opis, itd.
#plt.scatter(t, mag, s=3, label="Isto samo tačkice") # ako hoću tačkice, a ne liniju 
plt.legend() # ako želim da mi se štampa label koji sam napisala u plt.plot ili plt.scatter, ovo je važno kada imamo više od jedne krive na grafiku
plt.xlabel("t[dani]") # opis x-ose 
plt.ylabel("Magnituda") # opis y-ose
plt.grid() # daje mrežu u pozadini
#plt.title("Kriva sjaja") # naslov

#plt.gca().invert_yaxis() # invertuje osu 
```

```python
x = np.linspace(0, 2, 100)  # Sample data.

# drugi pristup, objekto orjentisan
# potpuno je svejedno da li definišete figuru ovako, ili kako smo ranije napisali, ali je lakše ako izaberete jedan od ova dva pristupa 
# koji hoćete da koristite, jer su neke komande malko različite (npr. plt.xlabel(), ali ax.set_xlabel(), i sl.)

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x, label='linear')  # Plot some data on the Axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the Axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the Axes.
ax.set_ylabel('y label')  # Add a y-label to the Axes.
ax.set_title("Simple Plot")  # Add a title to the Axes.
ax.legend()  # Add a legend.
```

```python
x = np.linspace(0, 2, 100)  # Sample data.

# isto kao iznad samo sa našim pristupom od ranije

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) Axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))

t = np.arange(0.0, 5.0, 0.01) # niz tačaka od 0 do 5 sa korakom 0.01
s = np.cos(2 * np.pi * t) # kosinusna funkcija
line, = ax.plot(t, s, lw=1) # lw mi je debljina linije

ax.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='red', shrink=0.05)) # čisto da znamo da i ovo postoji i da možemo da dodajemo tekst, strelice i sl. na grafik
            #ovde redom pišem tekst, poziciju centra strelice, poziciju teksta, i osobine strelice

ax.set_ylim(-2, 2) # mogu da ograničim y-osu (a mogu analogno ovome i x)
ax.set_xlabel('t')
ax.set_ylabel('cos(2pi*t)')
```

### Zadatak: Generišite 2000 random orbitalnih perioda egzoplaneta iz log-normal raspodele (možete koristiti funkciju iz numpy za to), sa srednjom vrednošću od 2 dana i sigmom od 1. Prikažite histogram ove raspodele.

### Zadatak: Postoje različite metode otkrivanja egzoplaneta, i neka je 5200 otkriveno zahvaljujući tranzitima, 1100 metodom radijalnih brzina, 80 uz pomoć Imaging metode, a 200 zahvaljujući mikrolensingu. Napraviti pie-chart koji prikazuje ovo.

```python
methods = ["Transit", "Radial Velocity", "Imaging", "Microlensing"] 
#definišem nazive koje želim na imam na pie-chart

counts = [5200, 1100, 80, 200] #naravno, brojevi moraju da budu istim redom kao i metode
#prethodno definisane da bi odgovarali jedni drugima
# ovde vidimo da counts može da bude i lista, ne mora da bude np niz :)

plt.figure() #treba da se koristi ako koristite .py na početku tj. pri definisanju figure
plt.pie(
    counts,
    labels=methods,
    autopct='%1.1f%%'
)

plt.title("Detekcija egzoplaneta")
#plt.show() #ovo može da bude potrebno ako koristite Notebooks i kuca se na kraju
```

```python
# orbita egzoplanete oko centralne zvezde
theta = np.linspace(0, 2*np.pi, 500) 

# velika i mala poluosa 
a = 7
b = 3

# prebacujem u pravougle koordinate
x = a*np.cos(theta)
y = b*np.sin(theta)

plt.plot(x,y)

plt.scatter(0,0, label="Star") # ovako npr. možemo da plotujemo samo jednu tačku!

plt.axis("equal") # korisno da bi skale na x- i y-osi bile jednake 
# ako probamo bez ovoga, izgleda dosta manje "elipsasto", što može da bude misleading
plt.legend()
```

### Zadatak: Dati su podaci za B-V, V i njihove neodređenosti u dokumentu "M12_podaci.txt". Nacrtajte CM dijagram globularnog jata M12 (zavisnost V magnitude od indeksa boje B-V) i dodajte error bars.

```python
data = np.loadtxt("M12_podaci.txt")

bv = data[:,0]
bv_err = data[:,1]

V = data[:,2]
V_err = data[:,3]

# alternativno, mogu da učitam podatke i ovako
# bv, bv_err, V, V_err = np.loadtxt("M12_podaci.txt", unpack=True, skiprows=1)

plt.figure(figsize=(7,8))

plt.scatter(bv, V, s=8, alpha=0.6) # s je size, alpha je prozračnost

plt.errorbar(
    bv,
    V,
    xerr=bv_err,
    yerr=V_err,
    fmt='none',
    alpha=0.5
)
# ovo koristim da bih prikazala neodređenosti, 

plt.gca().invert_yaxis() #manja magnituda - veći sjaj, veća mag - manji sjaj
# invertujem y osu

plt.xlabel("B - V")
plt.ylabel("V")

plt.title("CMD")

plt.grid(alpha=0.4)
```

```python
x = np.linspace(-5,5,200)
y = np.linspace(-5,5,200)

X, Y = np.meshgrid(x,y)

Z =np.exp(-(X**2 + Y**2))

plt.imshow(Z, origin='lower', cmap='gray')
# imshow je uglavnom koristan kada prikazujemo snimke sa teleskopa 

plt.colorbar(label="Sjaj") # prikazuje skalu boje, koja nosi jako bitnu informaciju!

plt.title("Fake Star")
```

```python
Z = (
    np.exp(-((X+2)**2 + Y**2))
    +
    0.7*np.exp(-((X-2)**2 + (Y-1)**2))
)

plt.imshow(Z, origin='lower', cmap='magma')

plt.colorbar(label="Intensity")
```

```python
plt.contour(X, Y, Z) #ovde takođe, kao i iznad imamo neki 2D data visualization, koji može biti koristan u nekim probabilstičkim analizama
plt.colorbar()
plt.title("Contour Plot")
```

```python
import seaborn as sns
# biblioteka zasnovana na matplotlib, može biti zanimljiva jer ima dosta lepe teme
sns.set_style("darkgrid") # možete da odaberete temu, i onda da svi grafici izgledaju u istom stilu, bez da morate manuelno to da podešavate
# tako da može biti korisno za pravljenje nekih postera, prezentacija, itd.
sns.scatterplot(x=bv, y=V, s=10)
```

```python
data = np.loadtxt("M12_podaci.txt")

bv = data[:,0]
V = data[:,2]

sns.set_theme(style="darkgrid")

# isto kao prethodno, samo korsiteći seaborn
sns.scatterplot(
    x=bv,
    y=V,
    s=10,
    alpha=0.5
)

#kao što vidite, sve ostale stvari na grafiku dodajemo isto kao i ranije, koristeći plt
plt.gca().invert_yaxis()

plt.xlabel("B - V")
plt.ylabel("V")

plt.title("CMD")
```

```python
# umesto tačkica, mogu da plotujem i raspodelu gustine, ovo je možda malo manje korisno u ovom slučaju jer nemamo neku skalu koja
# nam objašnjava šta ovo sve znači
sns.kdeplot(
    x=bv,
    y=V,
    fill=True,
    cmap="mako"
)

plt.gca().invert_yaxis()

plt.xlabel("B - V")
plt.ylabel("V Magnitude")


plt.title("CMD Density Map")
```
