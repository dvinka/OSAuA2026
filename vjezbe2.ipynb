```python
t = np.array([0, 1, 2, 3, 4]) #[s]
v = 5 #[m/s]
s = v*t #računam pređeni put za ravnomjerno pravolinijsko kretanje 
```

```python
import matplotlib.pyplot as plt #biblioteka koju koristim za prikazivanje podataka

plt.figure() #ovo zapravo samo pravi novu figuru
plt.scatter(t, s) #plotujem tačkice, prvo kažem šta je na x-, a onda na y- osi
plt.plot(t, s) #plotujem liniju
plt.xlabel('t[s]') #oznaka x-ose
plt.ylabel('s[m]') #oznaka y-ose
plt.title('Krećem se') #naslov
```

Zadatak za vas: Slobodan pad, hoću da vidim kako se visina mijenja u toku vremena i da pokažem grafik zavisnosti te visine od vremena.

```python
import matplotlib.pyplot as plt 

g = 9.81 # [m/s²]
t = np.linspace(0, 5, 100) #[s] #ovo mi daje 100 brojeva između 0 i 5, takvih da je razlika svaka dva susjedna člana niza jednaka

y0 = 500 #[m]

y = y0 - 0.5 * g * t**2 

plt.figure(figsize=(12,8))
plt.plot(t, y)
plt.xlabel("Vreme [s]")
plt.ylabel("Visina [m]")
plt.title("Slobodan pad")
plt.grid() #prikazuje mrežu u pozadini
```

Ako imam kosi hisac, koji izbacujem nekom početnom brzinom v0, kako se mijenja maksimalni pređeni put hica, u zavisnosti od ugla pod kojim je izbačen.

```python
v0 = 300 #[m/s]

theta = np.linspace(0, 90, 100) #uzimam uglove između 0 i 90, njih 100, isto kao iznad za vrijeme
theta = np.deg2rad(theta) #prebacujem ih u radijane, zato što np.sin (i ostale np trigonometrijske funkcije) rade sa radijanima

D = v0**2*np.sin(2*theta)/g #formula za maksimalan pređeni put kosog hica u zavisnosti od ugla

plt.figure(figsize=(12,8)) #ovde osim što kažem da hoću novi grafik, definišem i njegovu veličinu
plt.plot(np.rad2deg(theta), D) #x-osu sada pretvaram iz radijana u stepene, jer mi je tako prirodnije
plt.xlabel('Ugao [stepeni]')
plt.ylabel('Pređeni put [m]')
```

Po Keplerovom zakonu važi, 

$P^2 = a^3$ 

ako je P[god], a[AU]. Hajde da prikažemo tu zavisnost.

```python
a = np.arange(0.5, 10, 0.5) #ovde uzimam brojeve od 0.5 do 10, sa korakom od 0.5
P = np.sqrt(a**3)

plt.figure()
plt.plot(a, P)
plt.xscale('log') #ovde kažem hajde da mi x-skala bude prikazana logaritamski
plt.yscale('log') #isto i sa y -> onda vidim linearnu zavisnost!
plt.xlabel('a[AU]')
plt.ylabel('P[yr]')
```

```python
x, y = np.loadtxt('spektar.txt', usecols=(0,1), unpack=True) #učitavam podatke iz fajla
#potrebno je da fajl bude u istom folderu kao i .py, ili ako nije da napišem putanju do fajla 
#usecols je koristan zato što npr. mogu da imam 20 kolona, a ja hoću da učitam samo dvije; i ovde brojimo od nule!
#unpack kaže da hoću da ga razdvojim po kolonama, tj. raspakujem ;)
#delimiter između kolona je po defaultu tab, ali npr. ako su podaci u kolonama razdvojene zarezom , potrebno je dodati i delimiter=','

x = x[np.where(x<1500)] #ovde isijecam one podatke koji su mi zapravo spektralna linija (odokativno)
y = y[np.where(x<1500)] #ovde isto to radim, ali je bitno primjetiti da u oba slučaja kažem tamo gdje je x<1500

plt.figure()
plt.plot(x, y)
plt.xlabel('Talasna dužina')
plt.ylabel('Fluks')
```
