##### **Python is a high-level, interpreted, dynamically typed, garbage-collected, object-oriented, multi-paradigm programming language with strong support for procedural, functional, and object-oriented programming, and a huge standard library.**

##### A šta znače sve ove riječi? 
- high-level: nekako liči na jezik koji ljudi koriste (human-readable) 
- interpreted: kad pokrenemo program, on čita i radi liniju po liniju 
- dynamically typed: automatski dodjeljuje tip podatka promjenljivoj (o ovome više na sljedećem času) 
- garbage-collected: automatski upravlja memorijom 
- object-oriented: podržava klase/objekte (o ovome malo više nekad na kraju kursa) 
- multi-paradigm: podržava različite programske paradigme (paradigma = različiti načini organizovanja koda)  
- standard library: built-in alati
